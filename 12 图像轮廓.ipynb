轮廓是什么？

边缘检测能够测出边缘，但是边缘是不连续的

將边缘连接为一个整体，构成轮廓

注意问题：

1）处理的对象是二值图像，所以如果是灰度图像或彩色图像，需要先转换为二值图像，即需要预先进行阈值分割或边缘检测

2）查找轮廓需要更改原始图像，因此，通常需要拷贝一份原始图像，在拷贝的图像上进行查找操作，避免对原图像造成破化

3）在opencv中，是从黑色背景中查找白色对象，因此，对象必须是白色的，背景必须时黑色的，如果不是，需要预先转换

查找函数：

cv2.findContours() #查找图像轮廓

cv2.drawContours() #绘制对象轮廓

格式：

image,contours,hierarchy=cv2.findContours(image,mode,method)

image:修改为二值图像后的图像

contours:轮廓

hierarchy:图像的拓扑信息(轮廓层次）

image:原始图像

mode:轮廓检索模式

1)cv2.RETR_EXTERNAL 表示只检测外轮廓

2）cv2.RETR_LIST 检测的轮廓不建立等级关系

3）cv2.RETR_CCOMP 建立两个等级的轮廓，外面的一层为外边界，里面的一层外内孔的边界信息。如果内孔内还有一个连通物体，这个物体的边界也在顶层

4)cv2.RETR_TREE 建立一个等级树结构的轮廓 最常用

method：轮廓的近似方法，一般只用前两种

1）cv2.CHAIN_APPROX_NONE 存储所有的轮廓点，相邻的两个点的像素位置差不超过1

2）cv2.CHAIN_APPROX_SIMPLE 压缩水平方向、垂直方向、对角线方向的元素,只保留该方向的终点坐标，例如一个矩形轮廓只需4个点来保存轮廓信息

3）cv2.CHAIN_APPROX_TC89_L1 使用teh-Chinl chain近似算法

4）cv2.CHAIN_APPROX_TC89_KCOS 使用teh_Chinl chain近似算法


格式：

r=cv2.drawContours(img,contours,contourldx,color[,thickness])

r:目标图像，直接修改目标的像素点，实现绘制

img：原始图像

contours:需要绘制的边缘数组，从cv2.findContours()求出

contourldx:需要绘制的边缘索引，从0开始，如果全部绘制则为-1

color：绘制的颜色，为BGR格式Scalar，如[255,0,0]

thickness：可选，绘制的宽度，即描绘轮廓时所用的画笔粗细

```python
import cv2
img=cv2.imread("./image/contours2.bmp")
gray=cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)
ret,binary=cv2.threshold(gray,127,255,cv2.THRESH_BINARY)
#image,contours, hierarchy = cv2.findContours(binary,cv2.RETR_TREE,cv2.CHAIN_APPROX_SIMPLE)
cv2.imshow("original1",img)
contours,hierarchy=cv2.findContours(binary,cv2.RETR_TREE,cv2.CHAIN_APPROX_SIMPLE)
cv2.imshow("original2",img)

r=cv2.drawContours(img,contours,-1,(0,0,255),1) #可以发现cv2.drawContours()原地修改了图像（实际是没有將图像的值修改，而是对读取
#图像后的原值进行了修改，所以img的值变了，路径"/home/lyc/Desktop/python+opencv/chapter13/contours2.bmp"下的值并没有改变
#如果不想改变，可以先用co=o.copy()拷贝一个副本,然后用co代替img
cv2.imshow("original3",img)

cv2.imshow("result",r)
cv2.waitKey()
cv2.destroyAllWindows()
```
